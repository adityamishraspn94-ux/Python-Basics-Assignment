## **Q2: Difference between List and Tuple in Python**

**Answer:**

| Feature | List                 | Tuple                  |
| ------- | -------------------- | ---------------------- |
| Mutable | Yes (can be changed) | No (cannot be changed) |
| Syntax  | `[ ]`                | `( )`                  |
| Speed   | Slower               | Faster                 |
| Example | `[1,2,3]`            | `(1,2,3)`              |

## **Q5: Write a program to check whether a number is even or odd**

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

## **Q8: Remove duplicate values from a list using a set**

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = list(set(numbers))
print(unique_numbers)
```
